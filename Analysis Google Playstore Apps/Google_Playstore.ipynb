```python
import pandas as pd
data_a=pd.read_csv("googleplaystore.csv")
data_a.head()
```

```python
data_b=pd.read_csv("googleplaystore_user_reviews.csv")
data_b.head()
```

```python
print(data_a.isnull().sum())
```

```python
data_a['Rating'].fillna(data_a['Rating'].mean(), inplace=True)
```

```python
print(data_a.isnull().sum())
```

```python
data_a.dropna(subset=['Type','Content Rating'], inplace=True)
```

```python
print(data_a.isnull().sum())
```

```python
print(data_a[data_a['Current Ver'].isnull()])
```

```python
data_a['Rating']=data_a['Rating'].astype(int)
```

```python
data_a.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(data_a['Rating'],bins=20,kde=True,color='salmon')
plt.title("Distribution of App Ratings")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(y='Category',data=data_a, order=data_a['Category'].value_counts().index,color='Green')
plt.title("counts of Apps Category")
plt.grid(True,alpha=0.3)
plt.show()
```

```python
sns.scatterplot(x='Rating', y='Installs', hue='Category',data=data_a)
plt.title('Relationship between Installs and Ratings')
plt.legend(loc='upper right',bbox_to_anchor=(2, 1))
plt.show()
```

```python
avg_rating_by_category =data_a.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print(avg_rating_by_category)
```

```python
most_installed_apps = data_a[['App','Installs']].sort_values(by='Installs',ascending=False).head(10)
print(most_installed_apps)
```

```python
top_genres = data_a['Genres'].value_counts().head(5)
print(top_genres)
```

```python
Higest_installs_apps = data_a.loc[data_a.groupby('Category')['Installs'].idxmax()]
print(Higest_installs_apps[['Category','App','Installs']])
```

```python
sns.heatmap(data_b.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
data_b=data_b.dropna()
```

```python
sns.heatmap(data_b.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
data_b['Sentiment'].unique()
```

```python
data_b['Sentiment']=data_b['Sentiment'].replace({
    'Positive':1,
    'Negative':0,
    'Neutral':2
})
```

```python
data_b.info()
```

```python
average=data_b.groupby('App')['Sentiment'].mean()
average.sort_values(ascending=False,inplace=True)
print(average)
```

```python
result=average[average>1].index.tolist()
print(result)
```

```python
print("Best Rated apps (count):",len(result))
```

```python
print(data_a.loc[data_a['App'].isin(result)])
```

```python
import matplotlib.pyplot as plt
import squarify

type_counts = data_a['Type'].value_counts()

plt.figure(figsize=(4, 2))
squarify.plot(sizes=type_counts.values, label=type_counts.index, alpha=.8)
plt.title("Treemap of App Types (Free vs. Paid)")
plt.axis('off')
plt.show()
```

```python
paid_apps = data_a[data_a['Type'] == 'Paid']

paid_category_counts = paid_apps.groupby('Category').size()

plt.figure(figsize=(12, 6))
sns.barplot(x=paid_category_counts.index, y=paid_category_counts, palette='viridis')
plt.xticks(rotation=90)
plt.title('Number of Paid Apps per Category')
plt.xlabel('Category')
plt.ylabel('Number of Paid Apps')
plt.grid(axis='y',linestyle='--',alpha=0.3)
plt.tight_layout()
plt.show()
```
